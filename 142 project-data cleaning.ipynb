### Drop columns

### Convert data type & extract year

DONE `host_since	`: extract year  

DONE `host_response_rate`, `host_acceptance_rate`:  string to float. '90%' to 0.9

DONE `host_is_superhost`, `host_has_profile_pic`, `host_identity_verified`, `has_availability`, `instant_bookable`: string to 1/0. 't' to 1; 'f' to 0. 

DONE `bathrooms_text`
`price`

```python
## string to float
listing['host_response_rate'] = listing['host_response_rate'].str.strip('%').astype("float") / 100.0
listing['host_acceptance_rate'] = listing['host_acceptance_rate'].str.strip('%').astype("float") / 100.0
```

```python
## string boolean to 1/0 dummies
listing = listing.replace({'t': 1, 'f': 0})
```

```python
## host_since extract year
listing['host_since_year'] = listing["host_since"].str.split("-", 1, expand=True)[0].astype("float")
```

```python
## bathrooms_text extract number and type
listing[['bathrooms_count', 'bathrooms_type']] = listing['bathrooms_text'].str.split(" ", 1, expand=True)
## encode 'Shared half-bath', 'Half-bath', and 'Private half-bath' as bathroom_counts = 0.5
listing['bathrooms_count'] = listing['bathrooms_count'].replace({'Shared': 0.5, 'Private': 0.5, 'Half-bath': 0.5})
```

```python
## Price convert to float
listing['price'] = listing['price'].str.extract('(\d+)')[0].astype("int")
```

### Drop original columns that have been converted to other columns

```python
drop = ['host_since', 'bathrooms_text']
listing = listing.drop(columns = drop, axis = 1)
```

### Amenities

```python
def amenity_dummies(df):
    df = df.copy()
    df['amenities'] = df['amenities'].str.lower()
    df.loc[df['amenities'].str.contains('tv'), 'tv'] = 1
    df.loc[df['amenities'].str.contains('video|netflix|roku'), 'video'] = 1
    df.loc[df['amenities'].str.contains('refrigerator'), 'refrigerator'] = 1
    df.loc[df['amenities'].str.contains('oven'), 'oven'] = 1
    df.loc[df['amenities'].str.contains('sound system'), 'sound system'] = 1
    df.loc[df['amenities'].str.contains('long term stays allowed'), 'long term stays allowed'] = 1
    df.loc[df['amenities'].str.contains('heating'), 'heating'] = 1
    df.loc[df['amenities'].str.contains('kitchen'), 'kitchen'] = 1
    df.loc[df['amenities'].str.contains('coffee'), 'coffee'] = 1
    df.loc[df['amenities'].str.contains('refrigerator|freezer'), 'refrigerator'] = 1
    df.loc[df['amenities'].str.contains('first aid kit'), 'first aid kit'] = 1
    df.loc[df['amenities'].str.contains('washer'), 'washer'] = 1
    df.loc[df['amenities'].str.contains('drier'), 'drier'] = 1
    df.loc[df['amenities'].str.contains('free parking|free street parking'), 'free parking'] = 1
    df.loc[df['amenities'].str.contains('workspace'), 'workspace'] = 1
    df.loc[df['amenities'].str.contains('luggage dropoff allowed'), 'luggage dropoff allowed'] = 1
    df.loc[df['amenities'].str.contains('parking'), 'parking'] = 1
    df.loc[df['amenities'].str.contains('security camera'), 'security camera'] = 1
    df.loc[df['amenities'].str.contains('lock on bedroom door'), 'lock on bedroom door'] = 1
    df.loc[df['amenities'].str.contains('outdoor furniture'), 'outdoor furniture'] = 1
    df.loc[df['amenities'].str.contains('elevator'), 'elevator'] = 1
    df.loc[df['amenities'].str.contains('backyard|garden'), 'backyard '] = 1
    df.loc[df['amenities'].str.contains('bbq grill|barbecue'), 'BBQ grill'] = 1
    df.loc[df['amenities'].str.contains('air conditioning'), 'Air conditioning'] = 1
    df.loc[df['amenities'].str.contains('indoor fireplace'), 'Indoor fireplace'] = 1
    df.loc[df['amenities'].str.contains('outdoor dining area'), 'Outdoor dining area'] = 1
    df.loc[df['amenities'].str.contains('fans'), 'fans'] = 1
    df.loc[df['amenities'].str.contains('balcony'), 'balcony'] = 1
    df.loc[df['amenities'].str.contains('gym'), 'gym'] = 1
    df.loc[df['amenities'].str.contains('closet|clothing storage'), 'Clothing storage'] = 1
    df.loc[df['amenities'].str.contains('breakfast'), 'Breakfast'] = 1
    df.loc[df['amenities'].str.contains('children’s books and toys'), 'Children’s books and toys'] = 1
    df.loc[df['amenities'].str.contains('outdoor furniture'), 'outdoor furniture'] = 1
    df.loc[df['amenities'].str.contains('heater|heating'), 'heater'] = 1
    df.loc[df['amenities'].str.contains('piano'), 'piano'] = 1
    df.loc[df['amenities'].str.contains('game'), 'game'] = 1
    df.loc[df['amenities'].str.contains('private hot tub'), 'Private hot tub'] = 1
    df.loc[df['amenities'].str.contains('bike'), 'bike'] = 1
    df.loc[df['amenities'].str.contains('pets allowed'), 'Pets allowed'] = 1
    
    df.loc[:, 'tv': 'Pets allowed'] = df.loc[:, 'tv': 'Pets allowed'].fillna(0)
    df = df.drop('amenities', axis = 1)
    return df
```

```python
listing = amenity_dummies(listing)
```

### Property Types

```python
def property_type_dummies(df):
    df = df.copy()
    
    df.loc[df['property_type'].str.contains('rental unit'), 'rental unit'] = 1
    df.loc[df['property_type'].str.contains('Private'), 'Private'] = 1
    df.loc[df['property_type'].str.contains('home'), 'home'] = 1
    df.loc[df['property_type'].str.contains('condo'), 'condo'] = 1
    df.loc[df['property_type'].str.contains('hotel'), 'hotel'] = 1
    df.loc[df['property_type'].str.contains('apartment'), 'apartment'] = 1
    df.loc[df['property_type'].str.contains('townhouse'), 'townhouse'] = 1
    df.loc[df['property_type'].str.contains('guest'), 'guest'] = 1
    df.loc[df['property_type'].str.contains('Shared'), 'Shared'] = 1
    df.loc[df['property_type'].str.contains('Entire'), 'Entire'] = 1
    df.loc[df['property_type'].str.contains('aparthotel'), 'aparthotel'] = 1
    df.loc[df['property_type'].str.contains('cottage'), 'cottage'] = 1
    df.loc[df['property_type'].str.contains('loft'), 'loft'] = 1
    
    df.loc[:, 'rental unit': 'loft'] = df.loc[:, 'rental unit': 'loft'].fillna(0)
    df = df.drop('property_type', axis = 1)
    
    return df
```

```python
listing = property_type_dummies(listing)
```
